# Clean the dataset

```python
df_important.loc[:, "smiles"] = df_important["smiles"].drop_duplicates()
```

```python
df_important
```

```python
df_important = df_important.dropna(subset=['smiles'])
```

```python
df_important
```

```python
df_important.loc[:, 'name'] = df_important['name'].str.lower()
```

```python
df_important.loc[:, "name"] = df_important["name"].drop_duplicates()
```

```python
df_important = df_important.dropna(subset=['name'])
```

```python
df_important
```

# Split the dataset

```python
train_df, test_df = train_test_split(df_important, test_size=0.10, random_state=42)
```

```python
train_df.to_csv('train.csv', index=False)
test_df.to_csv('test.csv', index=False)
```
