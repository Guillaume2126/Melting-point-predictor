```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("../Data/30 000 mp -- BradleyMeltingPointDataset.xlsx")
```

```python
# Let's keep only the important columns
df_important = df[["name", "smiles", "mpC"]]
```

```python
df_important.shape
```

```python
# Let's see the number of duplicates
df_important["smiles"].duplicated().sum()/len(df_important)
```

# Clean the dataset

```python
# Let's drop the duplicates
df_important.loc[:, "smiles"] = df_important["smiles"].drop_duplicates()
df_important.loc[:, 'name'] = df_important['name'].str.lower()
df_important.loc[:, "name"] = df_important["name"].drop_duplicates()

df_important
```

```python
# Let's drop the NaN values
df_important = df_important.dropna(subset=['smiles'])
df_important = df_important.dropna(subset=['name'])
df_important
```

# Save as CSV

```python
df_important.to_csv('../Data/Clean_dataframe.csv', index=False)
```
